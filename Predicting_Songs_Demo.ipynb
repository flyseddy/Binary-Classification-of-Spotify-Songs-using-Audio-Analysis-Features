```python
my_model = joblib.load('Best_Model_SVM_82.pkl')
```

```python
df_x_test = pd.read_csv('X_test.csv')
df_x_test_copy = df_x_test[:] # Keep a copy of X_test songs so we can see song names, artist, etc.
df_y_test = pd.read_csv('y_test.csv')
```

```python
# Select only the features we need for the model
df_x_test = df_x_test.iloc[:, 4:]
df_y_test = df_y_test.iloc[:, 1:]
```

```python
df_x_test.head()
```

```python
my_pipeline = joblib.load('pipeline.pkl')
```

```python
x_test_transformed = my_pipeline.transform(df_x_test)
```

```python
final_predictions = my_model.predict(x_test_transformed)
```

```python
from sklearn.metrics import accuracy_score
final_accuracy = accuracy_score(df_y_test, final_predictions)
print(f"{final_accuracy*100}% accuracy on the test set")
```

```python
df_x_test_copy['Prediction'] = final_predictions
df_x_test_copy['Actual'] = df_y_test
final_predictions_df = df_x_test_copy[['name', 'artist', 'Actual', 'Prediction']]
pd.set_option('max_rows', 200)
```
